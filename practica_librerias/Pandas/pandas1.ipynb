```python
import pandas as pd
import numpy as np

# Estructura de datos
# Series: Un conjunto de elementos en los cuales vamos a almacenar cualquier conjunto de datos
#   Tambien tiene una secuencia ordenada (0,1,2,3,4,5...)
# Dataframe: Conjunto de series, Estructura bidimensional con columnas que pueden ser tambien de cualquier tipo
```

```python
# Serie
naranjas  = pd.Series([4,2,9,6,10,200])
```

```python
print(naranjas)
```

```python
colores = pd.Series(['rojo','azul','verde','amarillo'])
print(colores)
```

```python
materias = pd.Series({
    'Matimaticas': 60,
    'Física': 120,
    'Lenguaje': 69
})
```

```python
#propiedades de una serie
# .size: Devuelve el número de elementos de la serie
# .index: Devuelve una lista con los nombres de las filas del df
# .dtype: Devuelve el tipo de datos de los elementos de la serie
numeros = pd.Series([1,2,3,4,5,6,7,8,9])
```

```python
numeros.size
numeros.index
# numeros.dtypes
```

```python
colores[1:4]
```

```python
# Series fundamentales
numeros.sum()
```

```python
numeros.min()
```

```python
# numeros.std()
numeros.describe()
```

```python
# filtrar materias
print(materias[materias > 90])
print(materias.sort_values())
print(materias.sort_index())
print(materias)
```

# Dataframes

```python
# Un dataframe es un conjunto de series
# data = diccionario
# index = filas
# columns = columnas
```

```python
# dataframe a partir de un diccionario
data = {
    'nombre': ['matias', 'coni','javier','cata'],
    'carrera': ['Informática','Odontología','Ingeniería','Literatura'],
    'email': ['matias@gmail.com','coni.K@hotmail.com','javier@gmail.com','cata@gmail.com']
}
```

```python
estudiantes_df = pd.DataFrame(data)
```

```python
estudiantes_df
```

```python
df = pd.DataFrame([['matias', 23], ['coni', 23], ['javier', 24], ['cata', 25]], columns=['Nombre', 'Edad'])
df
```
